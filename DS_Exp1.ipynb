```python
import pandas as pd
data = pd.read_csv('Titanic-Dataset.csv')
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
print("Few Rows:")
print(df.head())
```

```python
print("Information:")
print(df.info())
```

```python
df['Age']=df['Age'].fillna(method='ffill').fillna(method='bfill')
```

```python
print("After forwar and backward filling:")
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['Cabin']=df['Cabin'].fillna('Unknown',limit=10)
print(df.isnull().sum())
```

```python
df=df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Sex']=lb.fit_transform(df['Sex'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Fare'] = scaler.fit_transform(df[['Fare']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df[['Pclass', 'Sex', 'Age', 'SibSp']])
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
corr = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr)
```
